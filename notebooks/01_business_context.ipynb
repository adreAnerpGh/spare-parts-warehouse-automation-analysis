# Spare-Parts Warehouse Analysis & Automated Warehouse Transition

This notebook introduces the operational context, business objectives, warehouse structure, and analytical scope of the project.

The project combines two connected analysis layers:
- operational inventory and replenishment analysis
- automated warehouse transition and allocation analysis

## Business Context

The warehouse environment analysed in this project supports spare parts for industrial workshop equipment.

The original operational problem involved moving from a manual spare-parts warehouse structure toward an automated vertical warehouse system. This required understanding both the inventory profile and the physical warehouse constraints.

The analysis therefore considers:
- item movement frequency
- inventory value and criticality
- supplier and replenishment behaviour
- manual warehouse locations
- UDC/container types
- item dimensions, volume, and storage suitability
- automated tray and compartment compatibility

## Main Business Problems

The analyses focus on several warehouse and inventory-management challenges:

- high SKU volumes with uneven movement distribution
- concentration of picking activity in a limited number of items
- intermittent spare-parts demand
- slow-moving and non-moving inventory
- supplier lead-time variability
- manual warehouse space utilisation
- item accessibility and picking efficiency
- physical suitability for automated storage
- tray/compartment allocation constraints
- operational preparation for warehouse automation

## Scope of the Analysis

The analyses included in this repository focus on two connected areas.

### Operational and Inventory Analysis
- inventory classification and segmentation
- movement-frequency analysis
- ABC inventory-value analysis
- XYZ demand-variability analysis
- inventory structure and concentration
- safety stock and reorder point review
- replenishment workflow analysis
- operational KPI monitoring

### Warehouse Automation Analysis
- manual warehouse structure review
- UDC/container analysis
- item dimension and volume analysis
- automated warehouse feasibility
- tray and compartment compatibility
- tray allocation optimisation
- manual vs automated storage comparison
- warehouse transition evaluation

The project does not focus on:
- production planning
- transportation optimisation
- customer demand forecasting
- financial accounting processes

## Planned Analytical Workflow

1. Validate dataset structure and consistency
